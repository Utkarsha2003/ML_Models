```python

#numericaldata
#categoricaldata
#ordinal
import numpy
x= [99,86,87,88,111,94,78,77,85,86,86]
mean= numpy.mean(x)
print('MEAN =',mean)

median = numpy.median(x)
print('MEDIAN =',median)

from scipy import stats
mode = stats.mode(x)
print('MODE =',mode)

standard_deviation= numpy.std(x)
print('STD= ',standard_deviation)

var=standard_deviation**2
variance= numpy.var(x)
print('Variance=',variance)
print('Variance=',var)
```

```python
percentile=numpy.percentile(x,9)
print('percentile =',percentile)
```

```python
#9% of the students scored less than or equal to the value 77.9
percentile=numpy.percentile(x,50)
print('percentile =',percentile)
```
