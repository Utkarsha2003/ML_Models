```python
# Simple Linear Regression

# Importing the libraries
import numpy as np
# numpy is used for handling arrays and numerical operations.
import matplotlib.pyplot as plt
# matplotlib.pyplot is used for plotting graphs and visualizations
import pandas as pd
# pandas is used for data manipulation and analysis, such as reading datasets and working with DataFrames.

# Set numpy display options to avoid scientific notation
np.set_printoptions(precision=2, suppress=True)
# This ensures that numpy arrays are printed with 2 decimal places and prevents scientific notation (suppress=True).
```

```python
# Importing the dataset
dataset = pd.read_csv(r"C:\Users\user\Salary_data.csv")
D = dataset.iloc[:, :].values
# all rows all cols
X = dataset.iloc[:, :-1].values  
# independent variable,all rows and all columns except the last one.
y = dataset.iloc[:, -1].values
# dependent variable,all rows from the last column
```

```python
print(D,"\n")
print("Independent-\n",X,"\n")
print("Dependent-\n",y,"\n")
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 1)
# test_size = 1/3: 1/3 of the data is allocated for testing, and the remaining 2/3 is used for training.
```

```python
# Training the Simple Linear Regression model on the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
# LinearRegression() creates a Linear Regression model.

regressor.fit(X_train, y_train)
# trains the model using the training data. It learns the relationship between Years of Experience (X_train) and Salary (y_train).
```

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test)
# predict(X_test) uses the trained model to make predictions on the test data (X_test), predicting Salaries based on Years of Experience in X_test.

print(y_pred)
```

```python
y_pred_train = regressor.predict(X_train)
# predicts the Salaries for the training data (X_train), and stores them in y_pred_train.

print(y_pred_train)
```

```python
# Visualising the Training set results
plt.scatter(X_train, y_train, color = 'red')
# This creates a scatter plot of the training data (Years of Experience vs Salary) in red.

plt.plot(X_train, regressor.predict(X_train), color = 'blue')
# This plots the regression line (predicted salary vs years of experience) in blue.

plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_test, y_test, color = 'red')
# Creates a scatter plot of the test data (Years of Experience vs Actual Salaries) in red.

plt.plot(X_train, regressor.predict(X_train), color = 'blue')
# plots the same regression line (from training data) over the test data to show how well the model generalizes to unseen data.

plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
