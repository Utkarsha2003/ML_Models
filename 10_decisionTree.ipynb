```python
# The Decision Tree algorithm is a popular supervised learning technique used for classification and regression tasks. When using entropy as the criterion, the algorithm splits data based on information gain, which measures the reduction in uncertainty (impurity) after a split.
```

```python
# Importing necessary libraries
import numpy as np
# numpy is used for handling arrays and numerical operations.
import matplotlib.pyplot as plt
import pandas as pd
# pandas is used for data manipulation and analysis, such as reading datasets and working with DataFrames.
```

```python
# Loading the dataset
dataset = pd.read_csv(r"C:\Users\user\Social_Network_Ads.csv")
X = dataset.iloc[:, :-1].values
# age,salary(independent)
y = dataset.iloc[:, -1].values
# purchased(dependent)
```

```python
# Splitting the dataset into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
# fit_transform on X_train: Compute and apply scaling using training data.
X_test = sc.transform(X_test)
# transform on X_test: Apply the same scaling (computed from X_train) to ensure consistent preprocessing for both training and test data.
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
# Training the Decision Tree Classification model on the Training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)

# criterion='entropy':
# Specifies the function to measure the quality of a split in the tree.
# The "entropy" criterion uses information gain to split nodes, which is based on the concept of entropy from information theory. It measures the uncertainty or impurity in the data.

classifier.fit(X_train, y_train)
```

```python
# Predicting a new result
print(classifier.predict(sc.transform([[30,87000]])))  # Output: [0]
```

```python
print(classifier.predict(sc.transform([[50,0]])))  
```

```python
# Output: [1]
print(classifier.predict(sc.transform([[0,0]])))
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min()-10, stop=X_set[:, 0].max()+10, step=0.25),
                     np.arange(start=X_set[:, 1].min()-1000, stop=X_set[:, 1].max()+1000, step=0.25))
grid_points = np.array([X1.ravel(), X2.ravel()]).T
batch_size = 10000  
predictions = []

for i in range(0, len(grid_points), batch_size):
    batch = grid_points[i:i + batch_size]
    predictions.extend(classifier.predict(sc.transform(batch)))

predictions = np.array(predictions).reshape(X1.shape)
plt.contourf(X1, X2, predictions, alpha=0.75, cmap=ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c=[ListedColormap(('red', 'green'))(i)], label=j)

plt.title('Decision Tree Classification (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
X_set, y_set = sc.inverse_transform(X_test), y_test
# Generate a reduced-resolution meshgrid for the test set
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min() - 10, stop=X_set[:, 0].max() + 10, step=1),  # Adjust resolution
    np.arange(start=X_set[:, 1].min() - 1000, stop=X_set[:, 1].max() + 1000, step=100)  # Adjust resolution
)
grid_points = np.array([X1.ravel(), X2.ravel()]).T
batch_size = 10000  
predictions = []

for i in range(0, len(grid_points), batch_size):
    batch = grid_points[i:i + batch_size]
    predictions.extend(classifier.predict(sc.transform(batch)))

predictions = np.array(predictions).reshape(X1.shape)
plt.contourf(X1, X2, predictions, alpha=0.75, cmap=ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())


# Plot test points
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(
        X_set[y_set == j, 0], X_set[y_set == j, 1],
        color=ListedColormap(('red', 'green'))(i),  # Use color instead of c
        label=j
    )

plt.title('decision tree (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
